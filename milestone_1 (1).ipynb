```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("flights_sample_3m.csv")
```

```python
df['FL_DATE'] = pd.to_datetime(df['FL_DATE'], errors='coerce')
```

```python
df = df[(df['FL_DATE'].dt.year >= 2021) & (df['FL_DATE'].dt.year <= 2022)]
```

```python
delay_cols = {
    'DELAY_DUE_CARRIER': 'Carrier',
    'DELAY_DUE_WEATHER': 'Weather',
    'DELAY_DUE_NAS': 'NAS',
    'DELAY_DUE_SECURITY': 'Security',
    'DELAY_DUE_LATE_AIRCRAFT': 'Late Aircraft'
}

def get_delay_reasons(row):
    reasons = []
    for col, label in delay_cols.items():
        if col in row.index and row[col] > 0:
            reasons.append(label)
    return ", ".join(reasons) if reasons else "No Delay"

# ✅ Only if columns exist
available_cols = [c for c in delay_cols.keys() if c in df.columns]

if available_cols:  
    df['DELAY_REASON'] = df.apply(get_delay_reasons, axis=1)
    df = df.drop(columns=available_cols)
else:
    print("No delay columns found in the dataset.")

```

```python
print(df.columns.tolist())
```

```python
df.head()
```

```python
# Replace NaN in delay columns with 0
df['DEP_DELAY'] = df['DEP_DELAY'].fillna(0)
df['ARR_DELAY'] = df['ARR_DELAY'].fillna(0)
```

```python
# If DELAY_REASON is missing, mark as 'Unknown'
df['DELAY_REASON'] = df['DELAY_REASON'].fillna('Unknown')
```

```python
# 3️⃣ Standardize data types
df['FL_DATE'] = pd.to_datetime(df['FL_DATE'])   # ensure date format
df['CANCELLED'] = df['CANCELLED'].astype(int)   # 0/1 instead of float
df['DIVERTED'] = df['DIVERTED'].astype(int)     # 0/1 instead of float
```

```python
# 4️⃣ Remove duplicates
df = df.drop_duplicates()
```

```python
# 5️⃣ Optional: remove unrealistic values
# If delay is negative beyond scheduling buffer, clip to 0
df['DEP_DELAY'] = df['DEP_DELAY'].clip(lower=0)
df['ARR_DELAY'] = df['ARR_DELAY'].clip(lower=0)
```

```python
df.head()
```

```python
#memory optimization
def optimize_memory(df):
    print(f"Memory before: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")

    for col in df.columns:
        col_type = df[col].dtype

        if pd.api.types.is_integer_dtype(col_type):
            df[col] = pd.to_numeric(df[col], downcast="integer")
        elif pd.api.types.is_float_dtype(col_type):
            df[col] = pd.to_numeric(df[col], downcast="float")
        elif col_type == object:
            # Convert to category if less than 50% unique values
            if df[col].nunique() / len(df[col]) < 0.5:
                df[col] = df[col].astype("category")

    print(f"Memory after: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
    return df

# Run on your dataset
df_clean = optimize_memory(df)
```

```python
# Save cleaned dataset (make sure df is your cleaned DataFrame)
df.to_csv("flights_cleaned.csv", index=False)

# Optional: faster loading format
df.to_parquet("flights_cleaned.parquet", index=False)
```
